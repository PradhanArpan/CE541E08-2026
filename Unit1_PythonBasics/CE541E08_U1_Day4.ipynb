## Section 1 — The four basic data types

| Type | Example | Engineering use |
|---|---|---|
| `int` | `45` | Session number, fitting count |
| `float` | `0.013`, `105.4` | Manning's n, elevation, discharge |
| `str` | `"concrete"`, `"KRS"` | Material name, station code |
| `bool` | `True`, `False` | Flood alert, quality flag, pipe full? |

In real data files (CSV from IMD, CWC) every column has a type. Understanding types prevents calculation errors.

### ▶ Cell 1 — Four types in engineering context

````python
session_number  = 4
discharge_m3s   = 0.03142
station_name    = "Krishnarajasagara"
flood_alert     = False

print(f"session_number  = {session_number}  → {type(session_number).__name__}")
print(f"discharge_m3s   = {discharge_m3s}  → {type(discharge_m3s).__name__}")
print(f"station_name    = {station_name}  → {type(station_name).__name__}")
print(f"flood_alert     = {flood_alert}  → {type(flood_alert).__name__}")
````

## Section 2 — Type conversion

CSVs from IMD/CWC often store numbers as text. You must convert explicitly:
```python
int("45")       → 45
float("45.6")   → 45.6
str(45)         → "45"
"TRUE" == "TRUE" → True  (string comparison → bool)
```

### ▶ Cell 2 — Type conversion: reading from a CSV row

````python
# Simulate what arrives when reading a CSV file
raw_day      = "17"
raw_rainfall = "45.6"
raw_flag     = "TRUE"

print("Raw (all strings):")
print(f"  raw_day      = '{raw_day}'  type: {type(raw_day).__name__}")
print(f"  raw_rainfall = '{raw_rainfall}'  type: {type(raw_rainfall).__name__}")

# Convert
day_num     = int(raw_day)
rainfall_mm = float(raw_rainfall)
quality_ok  = raw_flag == "TRUE"

print("
After conversion:")
print(f"  day_num     = {day_num}   type: {type(day_num).__name__}")
print(f"  rainfall_mm = {rainfall_mm}  type: {type(rainfall_mm).__name__}")
print(f"  quality_ok  = {quality_ok}  type: {type(quality_ok).__name__}")

depth_m = rainfall_mm / 1000
print(f"
Rainfall depth = {depth_m} m")

# ─── INSTRUCTOR NOTE ───────────────────────────────────────────
# Demonstrate: print("45.6" + 10) → TypeError
# Then: print(float("45.6") + 10) → 55.6 ✓
# This is the most common bug when loading CSV data in Unit 4.
# ───────────────────────────────────────────────────────────────
````

### ▶ Cell 3 — String operations on station data

````python
station_raw = "  KRISHNARAJASAGARA  "
station     = station_raw.strip().title()
river       = "Cauvery"
year        = 2024
filename    = f"{river}_{station.replace(' ','')}_{year}_flow.csv"

print(f"Raw    : '{station_raw}'")
print(f"Clean  : '{station}'")
print(f"File   : {filename}")

# Parsing a CSV data line
line  = "2024-07-15,245.6,GOOD"
parts = line.split(",")
date  = parts[0]
flow  = float(parts[1])
flag  = parts[2] == "GOOD"
print(f"
Date: {date}, Flow: {flow} m³/s, Quality OK: {flag}")
````

### ▶ Cell 4 — Boolean quality control flag

````python
rainfall_mm  = 287.4
prev_day     = 12.3
sensor_ok    = True

exceeds_max  = rainfall_mm > 250.0
big_jump     = rainfall_mm > prev_day * 10
suspect      = exceeds_max and big_jump and sensor_ok

print(f"Rainfall     : {rainfall_mm} mm")
print(f"Exceeds 250? : {exceeds_max}")
print(f">10× prev?   : {big_jump}")
print(f"SUSPECT FLAG : {suspect}")
if suspect:
    print("⚠ Requires manual verification.")

# ─── INSTRUCTOR NOTE ───────────────────────────────────────────
# Real hydrological QC uses exactly this logic.
# In Unit 3 (NumPy) we will do this for thousands of values
# simultaneously using boolean arrays.
# ───────────────────────────────────────────────────────────────
````
