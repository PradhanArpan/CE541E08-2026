## Section 1 — f-string format specification

```python
{value : [fill][align][width][.precision][type]}
```
- `{x:.3f}` → 3 decimal places
- `{x:>12.4f}` → right-align in 12 chars, 4 dp
- `{x:+.1f}` → always show sign
- `{x:.1%}` → percentage (×100 automatically)
- `{x:,}` → thousands separator

### ▶ Cell 1 — Format specifications

````python
Q=0.034521; Re=98745.6; pct=0.6734; population=1250000

print(f"{'Station':<18} {'Flow (L/s)':>12} {'Status':^10}")
print("─"*42)
print(f"{'Krishnarajasagara':<18} {Q*1000:>12.2f} {'OK':^10}")
print(f"{'Mettur':<18} {128.90:>12.2f} {'OK':^10}")
print(f"{'Biligundlu':<18} {0.00:>12.2f} {'DRY':^10}")
print()
print(f"Reynolds   : {Re:,.0f}")
print(f"Population : {population:.2e}")
print(f"Efficiency : {pct:.1%}")
````

### ▶ Cell 2 — Monthly rainfall summary with anomaly

````python
monthly = {"Jun":134.2,"Jul":118.4,"Aug":112.7,"Sep":95.3,"Oct":71.2,"Nov":43.8}
normals = {"Jun":120.0,"Jul":110.0,"Aug":110.0,"Sep":90.0,"Oct":75.0,"Nov":40.0}

print("="*58)
print(f"  {'Month':<10} {'Actual':>8} {'Normal':>8} {'Anomaly':>10} {'%Normal':>8}")
print("─"*58)
for m in monthly:
    diff = monthly[m]-normals[m]; pct = monthly[m]/normals[m]*100
    mark = "↑" if diff>0 else "↓"
    print(f"  {m:<10} {monthly[m]:>8.1f} {normals[m]:>8.1f} {mark}{diff:>+9.1f} {pct:>7.0f}%")
total = sum(monthly.values()); norm_t = sum(normals.values())
print("─"*58)
print(f"  {'TOTAL':<10} {total:>8.1f} {norm_t:>8.1f} {total-norm_t:>+10.1f} {total/norm_t*100:>7.0f}%")
print("="*58)
````

### ▶ Cell 3 — Professional design calculation sheet

````python
import math
D_mm=400; n=0.013; S=0.0008; L=650.0; f=0.018; g=9.81; nu=1.007e-6
D=D_mm/1000; A=math.pi*(D/2)**2; P=math.pi*D; R=A/P
V=(1/n)*R**(2/3)*S**0.5; Q=V*A
h_f=f*(L/D)*(V**2/(2*g)); Re=V*D/nu

def row(label,val,unit=""): print(f"  {label:<28} {val:>12}  {unit}")

print("="*52)
print(f"  {'PIPE FLOW DESIGN CALCULATION':^48}")
print("="*52)
print("
  INPUTS")
print("─"*52)
row("Pipe diameter",f"{D_mm} mm","")
row("Manning's n",f"{n}","—")
row("Slope",f"1 in {int(1/S)}","m/m")
print("
  RESULTS")
print("─"*52)
row("Area A",f"{A:.4f}","m²")
row("Velocity V",f"{V:.3f}","m/s")
row("Discharge Q",f"{Q:.4f}","m³/s")
row("Discharge Q",f"{Q*1000:.2f}","L/s")
row("Head loss",f"{h_f:.3f}","m")
row("Reynolds number",f"{Re:,.0f}","—")
print("
  CHECKS")
print("─"*52)
row("Velocity 0.6–3.0 m/s","PASS ✓" if 0.6<=V<=3.0 else "FAIL ✗","")
row("Flow regime","Turbulent ✓" if Re>4000 else "Laminar","")
print("="*52)
````
