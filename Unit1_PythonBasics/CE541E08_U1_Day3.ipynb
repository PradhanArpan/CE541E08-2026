## Section 1 — Variables: named containers

`pipe_diameter = 0.3` means: *store 0.3 in a container labelled pipe_diameter.*
The `=` is **assignment**, not equality.

**Good naming convention for engineering:**
| ❌ Bad | ✅ Good |
|---|---|
| `d` | `pipe_diameter` |
| `n` | `manning_n` |
| `q` | `discharge_m3s` |
| `x1` | `rainfall_mm` |

### ▶ Cell 1 — Creating and updating variables

```python
pipe_diameter = 0.3
print("Initial:", pipe_diameter, "m")

pipe_diameter = 0.6           # update — old value gone
print("Updated:", pipe_diameter, "m")

import math
area = math.pi * (pipe_diameter/2)**2
print("Area for new diameter:", round(area,4), "m²")

# ─── INSTRUCTOR NOTE ───────────────────────────────────────────
# Ask: "In Excel, what is the equivalent of a variable?"
# Answer: a named cell — same idea, different syntax.
# ───────────────────────────────────────────────────────────────
```

### ▶ Cell 2 — Engineering variable naming in Manning's equation

```python
import math

# Well-named variables — self-documenting code
pipe_diameter_m   = 0.3
manning_n         = 0.013
bed_slope         = 0.001

pipe_radius_m     = pipe_diameter_m / 2
flow_area_m2      = math.pi * pipe_radius_m**2
wetted_perim_m    = math.pi * pipe_diameter_m
hydraulic_rad_m   = flow_area_m2 / wetted_perim_m

velocity_ms       = (1/manning_n) * hydraulic_rad_m**(2/3) * bed_slope**0.5
discharge_m3s     = velocity_ms * flow_area_m2
discharge_lps     = discharge_m3s * 1000

print(f"Velocity   V = {velocity_ms:.3f} m/s")
print(f"Discharge  Q = {discharge_m3s:.5f} m³/s")
print(f"           Q = {discharge_lps:.2f} litres/s")

# ─── INSTRUCTOR NOTE ───────────────────────────────────────────
# Units in variable names (_m, _m2, _ms) prevent unit errors.
# A common engineering mistake: confusing m with mm.
# ───────────────────────────────────────────────────────────────
```

### ▶ Cell 3 — Multiple assignment patterns

```python
# Assign multiple variables on one line
width, depth, slope = 2.0, 0.8, 0.002
print(f"Width={width}m, Depth={depth}m, Slope={slope}")

# Swap two values without a temporary variable
upstream   = 105.4
downstream = 103.1
print(f"Before: up={upstream}, down={downstream}")
upstream, downstream = downstream, upstream
print(f"After:  up={upstream}, down={downstream}")

# Check types
print(f"
type(width)     = {type(width).__name__}")
print(f"type('concrete')= {type('concrete').__name__}")
print(f"type(True)      = {type(True).__name__}")
```

### ▶ Cell 4 — Augmented assignment: accumulating rainfall

```python
daily_rainfall = [12.4, 0.0, 35.6, 22.1, 8.9, 0.0, 41.2]
cumulative = 0.0

print(f"{'Day':<5} {'Daily (mm)':<14} {'Cumulative (mm)'}")
print("-" * 36)
for i, rain in enumerate(daily_rainfall, 1):
    cumulative += rain          # same as: cumulative = cumulative + rain
    print(f"{i:<5} {rain:<14.1f} {cumulative:.1f}")

print(f"
Weekly total: {cumulative:.1f} mm")

# ─── INSTRUCTOR NOTE ───────────────────────────────────────────
# += is the key pattern for running totals.
# Students will use this in Unit 2 for monthly rainfall sums.
# ───────────────────────────────────────────────────────────────
```
